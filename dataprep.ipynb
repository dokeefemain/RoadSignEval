```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
```

```python
df = pd.read_csv("lib/datasets/LISA/LISA/allAnnotations.csv",delimiter=";")
df
```

```python
names = df["Filename"].unique().tolist()
train, test = train_test_split(names,test_size=0.2)
train, val = train_test_split(train,test_size=0.25)
```

```python
with open("lib/datasets/train.txt", 'w') as file:
    file.write('\n'.join(train))
with open("lib/datasets/test.txt", 'w') as file:
    file.write('\n'.join(test))
with open("lib/datasets/val.txt", 'w') as file:
    file.write('\n'.join(val))
```
