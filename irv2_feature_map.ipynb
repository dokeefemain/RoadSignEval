```python
import tensorflow as tf
from PIL import Image
import numpy as np
from keras.models import Model
from matplotlib import pyplot as plt
```

```python
irv2 = tf.keras.models.load_model("lib/models/IRV2_Final1/IRV2_Final1")
```

```python
image_path = "lib/datasets/test_images/lisa_stop.png"
image = Image.open(image_path)
image1 = image.resize((299,299))
np_image = np.asarray(image1).astype('float64')
np_image /= 255
```

```python
model = Model(inputs=irv2.inputs,outputs=irv2.layers[-6].output)
model.summary()
```

```python
img1 = np_image.reshape((1, 299, 299, 3))
feature_maps = model.predict(img1)
plt.figure(figsize=(40,20))
ix = 1
for _ in range(4):
    for _ in range(8):
        ax = plt.subplot(4,8, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        plt.imshow(feature_maps[0, :, :, ix-1], cmap="gray")
        ix += 1
plt.savefig("lib/datasets/test_images/layer_"+"1"+".png")
plt.close()
```
